```python
import pandas as pd
```

## Series dans Pandas

Les Séries sont des tableaux unidimensionnels capables de stocker différents types de données. Elles sont particulièrement utiles pour représenter une colonne de données dans un ensemble de données tabulaire.


```python
#creation d'une serie
notes = pd.Series([17,20,15,16,14, 19])
notes
```

```python
# Note Max
notes.max()
```

```python
#Calcul de la moyenne
print('Moyenne des notes:', notes.mean())
```

## Data Frame dans Pandas

Les DataFrame sont des structures de données bidimensionnelles, similaires à des feuilles de calcul, où tu peux stocker et manipuler des données tabulaires avec des lignes et des colonnes.

```python
#creation de DF
data= {
    "Nom": ["Ibrahima", "Jibril", "Isham", "Aissatou"],
    "Age": [28, 16, 14, 23],
    "Ville": ["Abidjan", "Dakar", "Dakar", "Conakry"]
}

eleves = pd.DataFrame(data)
eleves
```

```python
#Moyenne d'age des élèves
moyenne_age = eleves["Age"].mean()
moyenne_age
```

```python
villes = eleves["Ville"]
villes
```

```python
# Ajouter une nouvelle note
eleves["Note"] = [19, 20, 20, 19]
eleves
```
