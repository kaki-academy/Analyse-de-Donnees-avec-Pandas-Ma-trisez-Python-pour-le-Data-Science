```python
import pandas as pd
```

```python
chemin = '/content/drive/MyDrive/Colab Notebooks/DataSets/Population_and_GDP_Africa_2022.csv'
donnees = pd.read_csv(chemin)
donnees
```

```python
donnees["GDP (USD)"] = donnees["GDP (USD)"].fillna(donnees["GDP (USD)"].mean())
donnees
```

```python
donnees.columns = donnees.columns.str.strip()
```

```python
pays_100 = donnees[donnees["Population"] > 100000000]
pays_100
```

```python
pib_popu = donnees[(donnees['GDP (USD)'] > 500000000) & (donnees['Population'] < 50000000)]
pib_popu
```

```python
critere = (donnees['Continent'] == 'West Africa') | (donnees['GDP (USD)'] / donnees['Population'] > 20000)
pays_critere = donnees[critere]
pays_critere
```

```python
pays_query = donnees.query('(`GDP (USD)` > 1e12) != (`Population` > 200000000) and Year == 2022 and Continent == "West Africa" ')
pays_query
```

```python
moyenne = donnees.groupby('Continent')['GDP (USD)'].transform('mean')
pib_sup = donnees[donnees['GDP (USD)'] > moyenne]
pib_sup
```

```python
donnees_eco = donnees.loc[donnees['Population'] > 100000000, ['Country', 'GDP (USD)', 'Population']]
donnees_eco
```

```python
colonnes_population = donnees.filter(like = 'Population')
colonnes_population
```
