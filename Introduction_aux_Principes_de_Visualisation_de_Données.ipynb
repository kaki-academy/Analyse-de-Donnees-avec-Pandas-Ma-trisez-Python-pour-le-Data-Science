```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#GRAPH LINEAIRE

annees = [2017, 2018, 2019, 2020, 2021]
ventes = [15000, 18000, 21000, 19500, 23000]

plt.plot(annees, ventes)
plt.xlabel('Annees')
plt.ylabel('Ventes(Euro)')
plt.title('Evolution annuelle des ventes')
plt.show()
```

```python
#GRAPH A BARRE

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/planets.csv')
sns.barplot(x='year', y='mass', hue='method', data=df)
plt.title('Mass of planet discovery by year and method')
plt.show()
```

```python
poids = [60,72,58,80,67,62]
taille = [170,182,168,185,175,173]

plt.scatter(poids, taille)
plt.xlabel('Poids(Kg)')
plt.ylabel('Taille(cm)')
plt.title('Rapport poids/taille')
plt.show()
```

```python
#HISTORGRAMME
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
sns.histplot(data=df, x='petal_width')
plt.title("Distribution de la largeur des ptales")
plt.show()
```

```python
slices = [35,5,30,20]
activites = ['travail', 'famille', 'sport', 'dormir']

plt.pie(slices, labels=activites, autopct="%1.1f%%")
plt.title('Rpartition de mon temps en journe')
plt.show()
```
