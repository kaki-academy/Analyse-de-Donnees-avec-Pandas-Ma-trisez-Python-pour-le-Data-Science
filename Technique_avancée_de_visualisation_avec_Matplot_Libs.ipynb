```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

```python
# Plot Interactif
x = list(range(1,11))
y = [10, 3, 6, 3, 2, 8, 3, 9, 5, 4]

# creer une figure
fig = go.Figure([go.Scatter(x=x, y=y)])

#creer un selector ou slider ou range
fig.update_layout(
    title='Selection iteractif avec Plotly',
    xaxis=dict(
        rangeselector= dict(
            buttons=list([
                dict(count=1, label='1m', step='month', stepmode='backward'),
                dict(count=6, label='6m', step='month', stepmode='backward'),
                dict(step='all'),
            ])
        ),
        type='date'
    )
)

fig.show()
```

```python
column_labels = ["Annee", 'Population', 'Taux de croissance']
data = [
    [2015, 120000, 1.3],
    [2016, 125600, 1.2],
    [2017, 130500, 1.4],
    [2018, 132800, 1.1],
]

fig, ax = plt.subplots()
ax.axis('off')

#Creation de notre tableau
tableau = ax.table(cellText=data, colLabels=column_labels, loc='upper left')

plt.title("Evilution Démographique")
plt.show()
```

```python
# Mise à jour de nos series
events = ['Lancement de produit', 'Expansion internationale', 'Introduction en bourse', 'Acquisition majeure']
years = [2005, 2010, 2015, 2020]

fig, ax = plt.subplots()
ax.plot(years,[1,1.2,1.3,1.1], marker="o", linestyle='--')
ax.set_xlabel('year')

y_positions = [1,1.2,1.3,1.1]

for year, event,y_pos in zip(years,events,y_positions) :
  ax.text(year, y_pos,event,ha='right')

plt.title('Timeline')
plt.show()
```
