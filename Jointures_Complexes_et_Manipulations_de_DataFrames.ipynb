```python
import pandas as pd
```

```python
# creation de DF

data = {
    "Nom": ['Ibrahima', 'Jean paul', 'Lionel', 'Boubacar'],
    "Age":  [28, 34, 29, 40],
    "Ville": ['New York', 'Paris', 'Berlin', 'London']
}

df = pd.DataFrame(data)

#trier par nom
df_nom = df.sort_values(by="Nom")

#trier par age
df_age = df.sort_values(by='Age', ascending=False )



```

```python
#Creation de DF2

data2 = {
    "Nom": ['Sylla', 'Moulod', 'Konan', 'Jibril'],
    "Age":  [30, 32, 31, 35],
    "Ville":  ['Atlanta', 'Chicago', 'Dallas', 'Seattle']
}

df2 = pd.DataFrame(data2)

#combiner df et df2
df_concact  = pd.concat([df, df2], ignore_index=True)
df_concact
```

```python
#Creation de DF3

data3 = {
   "Occupation": ['Engineer', 'Artist', 'Doctor', 'Writer'],
   "Salaire":  [90000, 75000, 110000, 85000]
}

df3 = pd.DataFrame(data3)
df_concact2 = pd.concat([df_concact, df3], axis=1)
df_concact2
```

```python
#Creation de DF4

data4 = {
    "Nom": ['Ibrahima', 'Jean paul', 'Lionel', 'Boubacar'],
    "Occupation": ['Engineer', 'Artist', 'Doctor', 'Writer'],
    "Salaire":  [90000, 75000, 110000, 85000]
}

df4 = pd.DataFrame(data4)

# Joindre df et df4
df_merge = pd.merge(df, df4, on='Nom')
df_merge
```

```python
#creation de df5

data5= {
    "Nom": ["Jean paul", "Lionel", "Marie", "Alexandre"],
    "Age" : [34, 28, 45, 36]
}

df5 = pd.DataFrame(data5)
```

```python
#jointure externe

df_outer = pd.merge(df4,df5, on='Nom', how='outer')
df_outer
```

```python
#Jointure gauche

df_left = pd.merge(df4, df5, on='Nom', how='left')
df_left
```

```python
df_right = pd.merge(df4, df5, on='Nom', how='right')
df_right
```
